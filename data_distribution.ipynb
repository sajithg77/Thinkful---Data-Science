```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Q1 Generate a random variable with 100 datapoints using the code..
binomial_dist = np.random.binomial(20, 0.2, 100)
```

```python
binomial_dist
```

```python
#Q2 Graph the variable using a histogram
plt.hist(binomial_dist)
plt.show()
```

```python
mean = np.mean(binomial_dist)
sd = np.std(binomial_dist)
```

```python
#3)Compute the mean and standard deviation and plot them as vertical lines on the histogram.
plt.hist(binomial_dist, color= 'g')
plt.axvline(binomial_dist.mean(), color='r', linestyle='solid', linewidth=2)
plt.axvline(binomial_dist.mean() + binomial_dist.std(), color='r', linestyle='dashdot', linewidth=2)
plt.axvline(binomial_dist.mean()-binomial_dist.std(), color='r', linestyle='dashdot', linewidth=2)
plt.show()
```

```python
### ADDITIONAL EXERCISE ####
# Generate two normally-distributed variables, one with a mean of 5 and standard deviation of 0.5, and the other with a mean of 10 and a standard deviation of 1.
n_dist1 = np.random.normal(5, 0.5, 100)
n_dist2 = np.random.normal(10, 1, 100)
```

```python
#Q2) Add them together to create a third variable.
n_dist3 = (n_dist1 + n_dist2)
```

```python
#Q3)Graph the third variable using a histogram.
plt.hist(n_dist3)
plt.show()
```

```python
#Q4)Compute the mean and standard deviation and plot them as vertical lines on the histogram.
mean = np.mean(n_dist3)
sd = np.std(n_dist3)
```

```python
plt.hist(n_dist3, color= 'black')
plt.axvline(x=mean,color='green')
plt.axvline(x=mean+sd, color='red')
plt.axvline(x=mean-sd, color='red')
plt.show()
```
