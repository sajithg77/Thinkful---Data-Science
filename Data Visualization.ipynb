```python
import pandas as pd
import numpy as np
import seaborn as sns
sns.set()
import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv('houseprice.csv')
```

```python
df.columns
```

```python
#1)Illustrate the distribution of saleprice.
# Highly concentrated 100k-200k
ax = sns.distplot(df['saleprice'], bins = 20)
```

```python
#2)Illustrate the distribution of the lotarea.
# without normalisation, we can say that the lotarea value is higly concentrated at 0.
ax1 = sns.distplot(df['lotarea'], bins =25, kde= False);
```

```python
#3)Use a boxplot to illustrate how overallcond affects the sale price.
# overallcond 9 has a high distribution in the upper quartile and 5 had the highest outliers
ax2 = sns.boxplot(y = 'saleprice', x ='overallcond', data=df)
```

```python
#4) On the same boxplot, highlight the difference between having central air conditioning or not.
# Its clear that central air had higher sales than non central air conditioning.
ax2 = sns.catplot(y = 'saleprice', x ='overallcond', col = 'centralair', kind='box' ,data=df)
```

```python
#5) Is there a correlation between lotarea and the saleprice?
#There is correlation. However, the value is as small as 26%.
df['lotarea'].corr(df['saleprice'])
```
