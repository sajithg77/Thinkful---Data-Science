```python
#1)Create a function that reverses a string. For example, given the string "forward", your function should return "drawrof". Assume all inputs are strings.
def reverse_letters (words):
    return words[::-1]
```

```python
#2) Complete the function below that can be used to calculate and return the posterior probability of an event A, given the known outcome of Event B using Bayes Theorem:

def bayes(a, b_given_a, b_given_not_a):
    return (b_given_a*a)/((a*b_given_a)+(1-a)*(b_given_not_a))
```

```python
bayes(0.4,0.3,0.8)
```

```python
#Q3) a) Analyze the teams dataset from the Lahman baseball records. This dataset contains season statistics for each Major League Baseball team.
df = pd.read_csv('lahman-teams.csv')
```

```python
df.columns
```

```python
df.hist('SB')
df.hist('ERA')
df.hist('W')
plt.show()
# Given their plots, "SB" is right skewed distribution,
#"W" is left skewed distribution, 
#"ERA" looks almost normally distributed.
```

```python
#b) Runs scored has a high correlation of 67.7 percent with winning.
df['R'].corr(df['W'])
```

```python
# The scaterplot clearly shows the increase in winning as the runs scored increases,
# Runs follow normal distribution.
sns.jointplot(x="W", y="R", data=df);
```

```python
# We can see from the trends that the errors occured in the game has been considerably minimised through the years.
sns.set_style("white")
s = df['E']
j = df['yearID']
plt.figure(figsize = (11,5))
plt.plot(j, s, "o", color='yellow')
plt.xlabel('Year ID')
plt.ylabel('Errors')
plt.title('Years vs Errors')
plt.show() 
```

```python
#Q4) Choose any two of the variables from 3. Draw 1,000 samples, taking the mean of 100 observations each time. Plot the frequencies of the sample means. What distribution does this appear to follow?
SB_means= []
for i in range(0,1000):
    ax1= df["SB"].sample(100)
    current_mean = ax1.mean()
    SB_means.append(current_mean)
    print(current_mean) 
```

```python
len(SB_means)
```
