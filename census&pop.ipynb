```python
import pandas as pd 
import numpy as np
```

```python
population_df = pd.read_csv('state-populations.csv')
census_div_df = pd.read_csv('census-divisions.csv')
```

```python
## 1) Merge these two DataFrames into a single DataFrame called census.
census_df = pd.merge(population_df, census_div_df)
census_df
```

```python
reshaped_df = pd.melt(census_df, value_vars=['2010', '2011', '2012', '2013', '2014', '2015', '2016'], var_name = 'year', value_name='population')
```

```python
##Q3 Re-shape census so that one column contains all population measures, and another the year attributes.
reshaped_df
```

```python
#3q: Group the data by year and summarize it. 
#ans: The population has increased with 2016 being the highest population. 
g  = reshaped_df.groupby('year').mean()
```

```python
g
```

```python
##q4 Group the data by region, division and year and summarize it.
##ans:The highest population is recorded at Mid Atlanta division and Northeast region. The lowest division is New England.
g1 = pd.melt(census_df, value_vars=['2010', '2011', '2012', '2013', '2014', '2015', '2016'], id_vars=['region','division'], var_name = 'year', value_name='population')
```

```python
g1.groupby(['region','division', 'year']).mean()
```
